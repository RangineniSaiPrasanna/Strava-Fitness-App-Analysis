<h1>Combined Data analysis</h1>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load individual datasets relevant for combination analysis
df_activity = pd.read_csv('/home/prasanna/Downloads/strava_csvfiles/dailyActivity_merged.csv')
df_calories = pd.read_csv('/home/prasanna/Downloads/strava_csvfiles/dailyCalories_merged.csv')
df_intensities = pd.read_csv('/home/prasanna/Downloads/strava_csvfiles/dailyIntensities_merged.csv')
df_steps = pd.read_csv('/home/prasanna/Downloads/strava_csvfiles/dailySteps_merged.csv')
df_sleep = pd.read_csv('/home/prasanna/Downloads/strava_csvfiles/sleepDay_merged.csv')
df_weight = pd.read_csv('/home/prasanna/Downloads/strava_csvfiles/weightLogInfo_merged.csv')
df_heartrate = pd.read_csv('/home/prasanna/Downloads/strava_csvfiles/heartrate_seconds_merged.csv')

# Convert date columns to datetime with explicit format
df_sleep['SleepDay'] = pd.to_datetime(df_sleep['SleepDay'], format="%m/%d/%Y %I:%M:%S %p")
df_weight['Date'] = pd.to_datetime(df_weight['Date'], format="%m/%d/%Y %I:%M:%S %p")
df_heartrate['Time'] = pd.to_datetime(df_heartrate['Time'])

# Convert activity-related dates
df_activity['ActivityDate'] = pd.to_datetime(df_activity['ActivityDate'])
df_calories['ActivityDay'] = pd.to_datetime(df_calories['ActivityDay'])
df_intensities['ActivityDay'] = pd.to_datetime(df_intensities['ActivityDay'])
df_steps['ActivityDay'] = pd.to_datetime(df_steps['ActivityDay'])

# Merge datasets
df_daily = df_activity.merge(df_calories, left_on=['Id', 'ActivityDate'], right_on=['Id', 'ActivityDay'], how='inner', suffixes=('_act', '_cal'))
df_daily = df_daily.merge(df_intensities, on=['Id', 'ActivityDay'], how='inner')
df_daily = df_daily.merge(df_steps, on=['Id', 'ActivityDay'], how='inner', suffixes=('', '_steps'))

print("Combined daily data shape:", df_daily.shape)
print(df_daily.head())

```

What is/are the insight(s) found from the chart?
